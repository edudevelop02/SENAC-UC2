```python
array_var = np.array(df['prazo_entrega_dias'])
array_quant = np.array(df['quantidade'])
array_valor_unitario = np.array(df['valor_unitario'])
print(array_quant)
print(array_valor_unitario)

media = np.mean(array_var)
mediana = np.median(array_var)
dist = (media - mediana) / mediana
print(f"Média: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Distância: {dist:.2%}")
print(f"quantidade: {dist:.2%}")
```
