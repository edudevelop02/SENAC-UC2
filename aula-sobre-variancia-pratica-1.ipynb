```python
'''
listar dados lendo uma base de dados externa
29/04/2026
'''
import pandas as pd
import numpy  as np

#dados = 'https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv'  # link na internet da base de dados externa
dados = 'BaseDPEvolucaoMensalCisp.csv'
df = pd.read_csv(dados, sep=';', encoding='latin-1')

#df
df.columns
```

```python
# Agrupando por CISP e somando os furtos de celular
df_cisp = df.groupby('cisp')['furto_celular'].sum().reset_index()
# Ordenando do maior para o menor
df_cisp = df_cisp.sort_values(by='furto_celular', ascending=False)
df_cisp
```

```python
# Usando .loc  para filtrar apenas os dados da capital  
# .groupby e .query() para filtrar os dados da capital
df_capital = df.loc[df['munic'] == 'Rio de Janeiro']
# Usando .query() â sintaxe alternativa mais legÃ­vel
df_capital = df.query("munic == 'Rio de Janeiro'")
# Repetindo o groupby apenas para a capital
df_cisp_capital = df_capital.groupby('cisp')['furto_celular'].sum().reset_index()
df_cisp_capital = df_cisp_capital.sort_values(by='furto_celular', ascending=False)
df_cisp_capital.head(5)
```

```python
# df_cisp_capital jÃ¡ foi criado no passo anterior (groupby + reset_index)
array_furto = np.array(df_cisp_capital['furto_celular'])
print(array_furto)
```

```python
# quartis
q1 = np.percentile(array_furto, 25)
q2 = np.percentile(array_furto, 50)
q3 = np.percentile(array_furto, 75)
print('Q1:', q1)
print('Q2:', q2)
print('Q3:', q3)
```

```python
# Usando .loc
df_top25 = df_cisp_capital.loc[df_cisp_capital['furto_celular'] > q3]
# Usando .query()
df_top25 = df_cisp_capital.query('furto_celular > @q3')
print(df_top25)
```

```python


# Desvio padrÃ£o das notas
desvio = np.std(df['roubo_veiculo'])
print(f"Desvio PadrÃ£o: {desvio:.2f}")
# SaÃ­da: Desvio PadrÃ£o: 1.74
# Interpretando com a mÃ©dia
media = np.mean(df['roubo_veiculo'])
print(f"MÃ©dia de roubo_veiculo: {media:.2f}")
print(f"Faixa tÃ­pica: {media-desvio:.2f} atÃ© {media+desvio:.2f}  ")
# SaÃ­da:
# MÃ©dia das notas: 7.06
# Faixa tÃ­pica: 5.32 atÃ© 8.80
```
